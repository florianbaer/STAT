```python
import pandas as pd
import matplotlib.pyplot as plt
import math
from scipy.stats import norm, binom
import scipy.stats as st

iron = pd.read_csv("ironF3.dat",sep=" ",index_col=False)
# print(iron.describe())
# print(iron.shape)
# print(iron)

for x in range(0,3):
    plt.subplot(2,3,x+1)
    iron.iloc[:,x].T.plot(kind="box")
    
#ironLog = iron.
for x in range(0,3):
    plt.subplot(2,3,x+4)
    iron.iloc[:,x].transform(lambda x: math.log(x)).T.plot(kind="box")
    
```
