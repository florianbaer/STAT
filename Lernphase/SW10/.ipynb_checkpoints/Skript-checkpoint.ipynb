```python
def boxcox(x,lambd):
    return np.log(x) if (lambd==0) else (x**lambd-1)/lambd

AirP["l_2"] = boxcox(AirP["Passengers"],2)
AirP["l_0"] = boxcox(AirP["Passengers"],0)
AirP["l_-05"] = boxcox(AirP["Passengers"],-.5)
plt.subplot(221)
AirP["Passengers"].plot()
plt.title("Original")
plt.xlabel("")
plt.subplot(222)
AirP["l_-05"].plot()
plt.title("lambda = -0.5")
plt.xlabel("")
plt.subplot(223)
AirP["l_2"].plot()
plt.title("lambda = 2")
plt.xlabel("")
plt.subplot(224)
AirP["l_0"].plot()
plt.title("lambda = 0")
plt.xlabel("")
plt.show()
```

```python
AirQ = pd.read_csv("AirQualityUCI.csv", sep = ";", decimal = ",")
AirQ1 = AirQ.copy()
# pandas kennt das Zeitformat in der Tabelle nicht:
# Punkt muss durch - ersetzt werden
AirQ1["Time"] = AirQ1["Time"].str.replace(".", ":")
AirQ1["Date"] = pd.to_datetime(AirQ1["Date"] + " " + AirQ1["Time"])
AirQ1.set_index("Date", inplace = True)
# Einige Werte der Temperatur sind -200. Diese Zeilen werden weggelassen
AirQ1 = AirQ1[AirQ1["T"] > -20]
AirQ1["T"].plot()
plt.show()
```

```python
AirQ1 = AirQ.copy()
# Einige Werte der Temperatur sind -200. Diese Zeilen werden weggelassen
AirQ1 = AirQ1[AirQ1["T"] > -20]
AirQ1["Time"] = AirQ1["Time"].str.replace(".", ":")
AirQ1["Date"] = pd.to_datetime(AirQ1["Date"] + " " + AirQ1["Time"])
AirQ1.set_index("Date", inplace = True)
# Resample to weekly frequency, aggregating with mean
AirQ1_weekly_mean = AirQ1['T'].resample('Y').mean()
AirQ1_weekly_mean.head(3)
print(AirQ1_weekly_mean.head(3))
```

```python
from pandas.plotting import lag_plot
AirQ4 = AirQ.copy()
AirQ4 = AirQ4[AirQ4["T"] > -20]
AirQ4["Time"] = AirQ4["Time"].str.replace(".", ":")
AirQ4["Date"] = pd.to_datetime(AirQ4["Date"] + " " + AirQ4["Time"])
AirQ4.set_index("Date", inplace = True)
plt.subplot(121)
lag_plot(AirQ4)
plt.subplot(122)
lag_plot(AirQ4, 10)
plt.show()
```
