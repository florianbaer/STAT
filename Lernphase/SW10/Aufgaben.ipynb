```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
pw_electric = pd.read_csv('pws.csv', sep=';', skiprows=2, header=0, encoding = "utf-8", index_col=0)
pw_electric.head()
pw_electric_zurich = DataFrame(pw_electric.ix["Zürich",1:])
pw_electric_luzern = DataFrame(pw_electric.ix["Luzern",1:])
pw_electric_luzern
pw_electric_zurich["Year"] = pd.to_datetime(pw_electric_zurich.index)
pw_electric_luzern["Year"] = pd.to_datetime(pw_electric_luzern.index)
pw_electric_luzern.set_index("Year", inplace=True)
pw_electric_zurich.set_index("Year", inplace=True)
log_cars1 = np.log(pw_electric_luzern.astype('float')) - np.log(pw_electric_luzern.shift(1).astype('float'))
log_cars = np.log(pw_electric_zurich.astype('float')) - np.log(pw_electric_zurich.shift(1).astype('float'))
log_cars1["Zurich"] = log_cars
log_cars1.head()
log_cars1.plot()
plt.xlabel("Jahr")
plt.show()
```

```python
def boxcox(x, lambd):
    return np.log(x) if (lambd == 0) else (x**lambd - 1) / lambd
```

```python
Electricity = pd.read_csv("AustralianElectricity.csv", sep=";", header=0)
Electricity.head()
Electricity["Quarter"] = pd.to_datetime(Electricity["Quarter"])
Electricity.set_index("Quarter", inplace=True)
Electricity.columns=["Electricity production Australia"]
Electricity.head()
Electricity2 = boxcox(Electricity, 0.2)
Electricity2.plot()
plt.ylabel("Million Kilowatthours")

```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
seasonal_decompose(Electricity2, model="additive", freq=4).plot()
plt.show()
```

```python
from stldecompose import decompose
for i in range(1,13):
    decompose(Electricity2, period=i).resid.plot()
    print(i)
    plt.show()
```
