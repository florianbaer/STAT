```python
from pandas import DataFrame, Series
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = Series([580.3, 568.2, 605.2, 659.1, 629.8, 322.3, 518.8, 600.4, 514.2, 576.2, 643.7, 575.8, 491.6])
```

```python
print(f'arithmetisches Mittel: {float(df.mean())}')
print(f'empirische Varianz: {float(df.var())}')
print(f'Standardabweichung: {float(df.std())}')
print(f'Median: {float(df.median())}')
```

```python
print(f'0.25% Quantil ist: {df.quantile(q=0.25)}')
print(f'0.50% Quantil ist: {df.quantile(q=0.50)}')
print(f'0.75% Quantil ist: {df.quantile(q=0.75)}')
```

```python
q25, q75 = df.quantile(q=[.25,.75])
print(f'Die Quantilsdifferenz ist: {q75 - q25}')
```

```python
# mit linspace siehts auch no interessant aus
q25, q50, q75 = df.quantile(q=np.linspace(start=0.25, stop=0.75, num=3))
print(f'0.25% Quantil ist: {q25}')
print(f'0.50% Quantil ist: {q50}')
print(f'0.75% Quantil ist: {q75}')
```

```python
df.plot(kind="hist", bins=7, edgecolor="black")
plt.title("Histogramm Tankfüllungen")
plt.xlabel("KM pro Füllung")
plt.ylabel("Häufigkeit")
plt.show()
```

```python
df.plot(kind="hist", bins=7, normed=True, edgecolor="black")
plt.title("Histogramm Tankfüllungen")
plt.xlabel("KM pro Füllung")
plt.ylabel("Häufigkeit")
plt.show()
```

```python
# Boxplot
df.plot(kind="box", title="Boxplot Tankfüllungen", label="Tankung")
plt.ylabel("KM pro Füllung")
plt.show()
```

```python
# empirische kumulative Verteilungsfunktion
df.plot(kind="hist", cumulative=True, histtype="step",
normed=True, bins=8, edgecolor="black")
```
