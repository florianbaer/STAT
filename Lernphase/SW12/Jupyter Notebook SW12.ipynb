# Mathematische Modellierung von Zeitreihen

### Random Walk

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
d = np.random.choice(a=[-1,1], size=10000, replace=True)
x = np.cumsum(d)
plt.plot(x)
plt.xlabel("Anzahl Schritte seit Start")
plt.ylabel("Abweichung von Startposition in Meter")
plt.show()
```

### Random Walk mit Drift

```python
np.random.seed(35)
d = np.random.choice(a=[-1,1], size=10000, replace=True)
delta = 5*10**(-2)
x = np.cumsum(d)
y = np.zeros(10000)
for i in range(1,10000):
    y[i] = delta+y[i-1]+d[i]
plt.plot(y)
plt.plot(x)
plt.xlabel("Anzahl Schritte seit Start mit Drift")
plt.ylabel("Abweichung von Startposition in Meter")
plt.show()
```

### Weisses Rauschen

```python
w = np.random.normal(size = 1000)
plt.plot(w)
plt.show()
```

### Moving Average Prozess

```python
from pandas import Series, DataFrame
w = DataFrame(np.random.normal(size=1000))
w.rolling(window=3).mean().plot()
plt.show()
```

### Autoregressiver Prozess

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.random.seed(35)
d = np.random.choice(a=[-1,1], size=500, replace=True)
y = np.zeros(500)
for i in range(2,500):
    y[i] = 1.5*y[i-1]-0.9*y[i-2]+d[i]
plt.plot(y)
plt.xlabel("Autoregressiver Prozess")
plt.ylabel("y-Abweichung")
plt.title("AR(2)-Prozess")
plt.show()
```

# Beispiel Autokorrelationsfunktion

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
df = pd.read_table("../../../Themen/Time_Series_mathematical_models/Skript_de/Daten/wave.dat")
print(df.head())
```

```python
%matplotlib inline

df.plot()

plt.xlabel("Time")
plt.ylabel("Wave height (mm)")

plt.show()
```

```python
df.loc[0:59,:].plot()

plt.xlabel("Time")
plt.ylabel("Wave height (mm)")

plt.show()
```

```python
from statsmodels.tsa.stattools import acf 

acf(df["waveht"])[1]
```

```python
from pandas.plotting import lag_plot
lag_plot(df,1)

plt.xlabel("x_t")
plt.ylabel("x_(t+k)")

plt.show()
```

```python
acf(df["waveht"])[[2,3,5,10]]
print(acf(df["waveht"])[[2,3,5,10]])
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(df, lags=25)

plt.show()
```
