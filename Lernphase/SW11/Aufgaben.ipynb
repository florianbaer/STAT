```python
n = 250
loc = 0.483 
scale = 11
from scipy.stats import norm
steps = np.array(norm.rvs(size = n, loc = loc, scale = scale))
sp_simulated = np.empty([250])
sp_simulated[0] =  1257.6
print(sp_simulated[0])
for i in range(249):
    sp_simulated[i+1] = sp_simulated[i]+ steps[i]
plt.plot(sp_simulated)
plt.xlabel("Simulierte Zeit")
plt.ylabel("Simulierte Werte")
plt.title('Weitere werte')
plt.show()
```
