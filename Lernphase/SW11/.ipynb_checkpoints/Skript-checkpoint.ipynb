```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
import pandas as pd
from scipy.stats import norm, uniform, expon, t, probplot
import scipy.stats as st
from scipy.integrate import quad
from sympy.solvers import solve
from sympy import Symbol
import numpy as np
from pandas import Series, DataFrame
```

```python
d = np.random.choice(a=[-1,1], size=10000, replace=True)
x = np.cumsum(d)
plt.plot(x)
plt.xlabel("Anzahl Schritte seit Start")
plt.ylabel("Abweichung von Startposition in Meter")
plt.show()
```

```python
np.random.seed(35)
d = np.random.choice(a=[-1,1], size=10000, replace=True)
delta = 5*10**(-2)
print(delta)
x = np.cumsum(d)
y = np.zeros(10000)
for i in range(1,10000):
    y[i] = delta+y[i-1]+d[i]
plt.plot(y)
plt.plot(x)
plt.xlabel("Anzahl Schritte seit Start mit Drift")
plt.ylabel("Abweichung von Startposition in Meter")
plt.legend(("drift", "no drift"))
plt.show()
```

```python
w = np.random.normal(size = 1000)
plt.plot(w)
plt.legend("Gauss'schen weisses Rauschen")
plt.show()
```

```python
w = DataFrame(np.random.normal(size=1000))
w.rolling(window=3).mean().plot()
plt.legend("Gauss'schen weisses Rauschen mit moving average")
plt.show()
```

```python
np.random.seed(35)
d = np.random.choice(a=[-1,1], size=500, replace=True)
y = np.zeros(500)
for i in range(2,500):
    y[i] = +1.5*y[i-1]-0.9*y[i-2]+d[i]
plt.plot(y)
plt.xlabel("Autoregressiver Prozess")
plt.ylabel("y-Abweichung")
plt.title("AR(2)-Prozess")
plt.show()
```
