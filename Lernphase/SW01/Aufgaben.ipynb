```python
from pandas import Series, DataFrame
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
geysir = pd.read_csv("geysir.dat", sep=" ", index_col=0)
```

```python
geysir.head()
```

```python
plt.subplot(221)
geysir["Zeitspanne"].plot(kind="hist", edgecolor="black")
plt.xlabel("10 Klassen")
plt.subplot(222)
geysir["Zeitspanne"].plot(kind="hist",
bins=20,
edgecolor="black")
plt.xlabel("20 Klassen")
plt.subplot(223)
geysir["Zeitspanne"].plot(kind="hist",
bins=np.arange(41,107,11),
edgecolor="black")
plt.xlabel("Klassengrenzen 41, 52 , 63, 74 , 85, 96")
plt.show()
```

```python
plt.subplot(221)
geysir["Eruptionsdauer"].plot(kind="hist", edgecolor="black", bins=5)
plt.xlabel("5 Klassen")
plt.subplot(222)
geysir["Eruptionsdauer"].plot(kind="hist",bins=10,edgecolor="black")
plt.xlabel("10 Klassen")
plt.subplot(223)
geysir["Eruptionsdauer"].plot(kind="hist",bins=15,edgecolor="black")
plt.xlabel("Klassengrenzen 41, 52 , 63, 74 , 85, 96")
plt.subplot(224)
geysir["Eruptionsdauer"].plot(kind="hist",bins=20,edgecolor="black")
plt.xlabel("Klassengrenzen 41, 52 , 63, 74 , 85, 96")
plt.show()
```
